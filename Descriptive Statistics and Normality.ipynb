# Descriptive Statistics and Normality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


poisson = np.random.poisson(7, 100)

plt.axvline(poisson.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean()-poisson.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean()+poisson.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(poisson)
plt.show()
```

For this poisson distribution, the probability distribution is skewed as it is asymmetric.  The mean is not clustered in the center of the data and the standard deviation is not comprised of the same amount of data on each side of the mean.

```python
norm1= np.random.normal(5, .5, 1000)
norm2= np.random.normal(10, 1, 1000)
norm = norm1 + norm2
plt.axvline(norm.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm.mean()-norm.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm.mean()+norm.std(), color='b', linestyle='dashed', linewidth=2)
plt.hist((norm), color='purple')
plt.show()
```

In this distribution, the data is symetrically distributed with the mean falling where the data clusters.  Also, the data is centered with approximately 50% of the data on each side of the mean.  Furthermore, approximately 67% of the data is within one standard deviation of the mean.
